#### 3) Find the Flats those are in resaling mode and remove the 7 Lac from target price also give the approval from ReRa

```python
data1=df.loc[(df['RESALE']==1)&(df['RERA']==1)]
data1.head()
```

```python

def new():
    df2=[]
    df3=[]
    for x in data1['TARGET(PRICE_IN_LACS)']:
        p=x-7
        df2.append(round(p,2))
    for y in data1['RERA']:
        r=y-1
        df3.append(r)
    data1.insert(12,'price_decrease by_7',df2)
    data1.insert(13,'New_Rera',df3)
    data1

new()
```

#### 5) Find the Ready to move Flats those are having ReRa Approved and Add the 4 Lac in Target price and Remove 3 Lac amount from Target Price those are not having Rera Approval

```python
df4=[]
data3=df.loc[(df['READY_TO_MOVE']==0)]
for x in data3['TARGET(PRICE_IN_LACS)']:
    for y in data3['RERA']:
        if y==0:
            p=x+4
            df4.append(p)
        else:
            p1=x-3
            df4.append(p1)
print(df4) 
```
