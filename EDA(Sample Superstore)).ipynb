### Handling Outliers

### Removing Outliers

```python
Q1 = np.percentile(data['Sales'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(data['Sales'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
Q1,Q3,IQR
```

```python
# Upper bound
upper=np.where(data['Sales']>=(Q3+1.5*IQR))
lower=np.where(data['Sales']<=(Q3-1.5*IQR))
```

```python
data.drop(upper[0], inplace = True)        #Removing the Outliers
data.drop(lower[0], inplace = True)
print('New Shape:',data.shape )
```

```python
sns.boxplot(x='Sales',data=data)
```

```python
sns.boxplot(x='Discount',data=data)
```

```python
Q1 = np.percentile(data['Discount'], 25,interpolation = 'midpoint')
  
Q3 = np.percentile(data['Discount'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1
print(Q1,Q3,IQR)
# Upper bound
upper=np.where(data['Discount']>=(Q3+1.5*IQR))
lower=np.where(data['Discount']<=(Q3-1.5*IQR))
data.drop(upper[0], inplace = True)        #Removing the Outliers
data.drop(lower[0], inplace = True)
print('New Shape:',data.shape)
sns.boxplot(x='Discount',data=data)
```
